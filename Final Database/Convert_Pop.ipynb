```python
# read csv file and convert into DF object
wcpop_df = pd.DataFrame(pd.read_csv("2021-standard-population-estimates.csv"))

wcpop_df.head(15)
```

```python
wcpop_df.dtypes
```

```python
wcpop_df = wcpop_df.rename(columns={'Municipality': 'PHYSICAL_CITY'})
wcpop_df = wcpop_df.rename(columns={'Geo Point': 'Geo_Point'})
wcpop_df.head(10)
```

```python
null_count = wcpop_df['Year Incorporated'].isna().sum()
print(null_count)
```

```python
columns_null = wcpop_df.isna().sum()
print(columns_null)
```

```python
wcpop_df = wcpop_df.fillna(0)
```

```python
wcpop_df = wcpop_df.astype({'Year Incorporated':'int'})
```

```python
wcpop_df.dtypes
```

```python
# remove Balance of Wake County string
wcpop_df = wcpop_df[wcpop_df['PHYSICAL_CITY'].str.contains("Balance of Wake County") == False]
wcpop_df = wcpop_df[wcpop_df['Estimate Type'].str.contains("2020 Base") == False]
wcpop_df = wcpop_df[wcpop_df['Estimate Type'].str.contains("Revised Estimate") == False]
```

```python
#wcpop_df = wcpop_df.drop(['County'])
#wcpop_df.head(5)
```

```python
# Converting Total_Sale_Date to datetime format
wcpop_df['Year'] = pd.to_datetime(wcpop_df['Year'], format ="%Y-%m-%d")
wcpop_df = wcpop_df.rename(columns={'Year': 'Estimate Date'})
# Checking to see if it worked
wcpop_df.dtypes
```

```python
# Seperating Total_Sale_Date into y m d columns
wcpop_df['year']= wcpop_df['Estimate Date'].dt.year
wcpop_df['month']= wcpop_df['Estimate Date'].dt.month
wcpop_df['day']= wcpop_df['Estimate Date'].dt.day

wcpop_df.dtypes
```

```python
# Seperating Geo Point into Latitude and Longitude
wcpop_df[['Latitude', 'Longitude']] = wcpop_df.Geo_Point.str.split(",", expand = True)
wcpop_df.head(10)
```

```python
# Drop Geo Point
wcpop_df = wcpop_df.drop(['Geo_Point', 'County'], axis=1)
#wcpop_df = wcpop_df.drop(['County'])
wcpop_df.head(5)
```

```python
# Change Lat and Long to float data type
wcpop_df = wcpop_df.astype({'Latitude':'float', 'Longitude':'float'})
wcpop_df.dtypes
```

```python
wcpop_df.to_csv ("v2Clean_WCpopulation-estimates.csv",
                  index = None,
                  header=True)
```
