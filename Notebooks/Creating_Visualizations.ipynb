CREATING VISUALIZATIONS FOR ALL CLEANED DATA

```python
# Scatter plot total Heated Area x Total Sale Price
plt.scatter(df['HEATED_AREA'], df['Total_sale_Price'])
plt.xlabel('Total Square Feet')
plt.ylabel('Total Sale Price')
plt.title('Total Sale Price vs Total Square Feet')
plt.show()
```

```python
# Group data by Year and Avg Sale price
house_price_by_year = df.groupby('year')['Total_sale_Price'].mean()

# Create a line plot to visualize the time series
plt.figure(figsize=(10, 6))
house_price_by_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Total Sale Price AVG')
plt.title('Time Series of House Sale Price by Year')
plt.show()
```

These two graphs above show the vase different to the little bit of EDA we performed before highlighting the amount of outliers. These graphs (especially the first one shows us how much more precise the new dataset is. 

```python
# Create a histogram showing heated area and deeded acreage
df.hist(column = ['HEATED_AREA', 'Deeded_Acreage'])
```

```python
# Checking relationship with Heated_Area and Total sale Price
df.plot.scatter(x='HEATED_AREA', y='Total_sale_Price', s=0.1, c = df.Year_Built,cmap='Blues')
```

There is a clear positive distribution between these two variables.
